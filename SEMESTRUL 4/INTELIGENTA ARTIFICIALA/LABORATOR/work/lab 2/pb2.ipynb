```python
import os
from PIL import Image
import matplotlib.pyplot as plt

image_folder = ("data/images/")
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

fig, axes = plt.subplots(1, len(image_files), figsize=(12, 4))
for i, file in enumerate(image_files):
    img = Image.open(os.path.join(image_folder, file))
    img_resized = img.resize((128, 128))
    axes[i].imshow(img_resized)
    axes[i].axis('off')
plt.show()
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

image_folder = "data/images"
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

fig, axes = plt.subplots(1, len(image_files), figsize=(12, 4))
for i, file in enumerate(image_files):
    img = Image.open(os.path.join(image_folder, file)).convert('L')
    img_array = np.array(img)
    img_inverted = 255 - img_array
    axes[i].imshow(img_inverted, cmap='gray')
    axes[i].axis('off')
plt.show()
```

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

image_path = "data/images/Russell.jpg"

img = Image.open(image_path)

img_blurred = img.filter(ImageFilter.GaussianBlur(radius=5))

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Before')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_blurred)
plt.title('After')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image_path = "data/images/Russell.jpg"

img = Image.open(image_path)

img_gray = np.array(img.convert('L'))

edges = cv2.Canny(img_gray, threshold1=100, threshold2=200)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Before')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('After')
plt.axis('off')

plt.show()
```
